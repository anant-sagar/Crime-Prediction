```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express
from pandas_profiling import ProfileReport
```

```python
df = pd.read_csv('datasets/communities-crime-clean.csv')
df['highCrime'] = np.where(df['ViolentCrimesPerPop']>0.1, 1, 0)
cols = ['communityname','state','fold']
df.drop(cols,axis=1,inplace=True)
```

```python
df.shape
```

```python
for col in df:
    ax, fig = plt.subplots(figsize=[25,5])
    ax = sns.countplot(x=col,data=df)
    ax.set_title(col.title()+" distrubution")
    xticks=ax.xaxis.get_major_ticks()
    for i in range(len(xticks)):
        if i%2==1:
            xticks[i].set_visible(False)
    plt.show()
    
```

```python
for col in df:
    sns.lmplot(x=col,y='ViolentCrimesPerPop',data=df)
    plt.show()
    break
```
